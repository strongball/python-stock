```python
import datetime
import backtrader as bt
import backtrader.feeds as feeds
import math
import yfinance as yf
# 從Yahoo Finance取得資料
data = feeds.PandasData(dataname=yf.download('TQQQ', '2015-07-06', '2021-07-01', auto_adjust=True))
```

```python
from MABadge import MABadge
```

```python
# 初始化cerebro
cerebro = bt.Cerebro()
# # feed data
cerebro.adddata(data)
# # add strategy
cerebro.addstrategy(MABadge)
print('Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())
# run backtest
cerebro.run()
print('Final Portfolio Value: %.2f' % cerebro.broker.getvalue())
# plot diagram
# cerebro.plot()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure()
cerebro.plot(iplot=False)
plt.show()
```
